```python
# Layoffs 2022-2024 Dataset Analysis
# Comprehensive SQL queries and visualizations in Google Colab

# Step 1: Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Step 2: Install necessary packages
!pip install pandas sqlalchemy ipython-sql matplotlib seaborn plotly

# Step 3: Import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from datetime import datetime
from IPython.display import display, HTML
import warnings
warnings.filterwarnings('ignore')

# Print title and description
print("# Layoffs Data Analysis (2022-2024)")
print("This notebook analyzes trends in tech industry layoffs using SQL queries in Google Colab")

# Step 4: Load the data directly from specified Google Drive path
try:
    df = pd.read_csv('/content/drive/My Drive/Data_Analyst/Portfolio_Projects/Projects/SQL/Layoffs_Dataset_Cleaned_20250406_134946.csv')
    print("✅ Data loaded successfully from Google Drive")
except Exception as e:
    print(f"❌ Error loading data: {str(e)}")
    print("Please ensure the file exists at the specified location.")
    # Exit or provide alternative if file doesn't exist
    raise

# Display initial dataset information
print("\n## Dataset Overview")
print(f"Dataset shape: {df.shape} (rows x columns)")
print("\nColumn names:")
for col in df.columns:
    print(f"- {col}")

print("\nData sample:")
display(df.head())

print("\nData types:")
display(df.dtypes)

print("\nMissing values:")
display(df.isnull().sum())

# Convert date column to datetime if it's not already
if df['date'].dtype != 'datetime64[ns]':
    df['date'] = pd.to_datetime(df['date'])
    print("✅ Date column converted to datetime format")

# Step 5: Create a SQLite database
from sqlalchemy import create_engine
import sqlite3

# Create a SQLite database
conn = sqlite3.connect('layoffs.db')
df.to_sql('layoffs_staging2', conn, if_exists='replace', index=False)
print("✅ SQLite database created and data loaded")

# Step 6: Setup SQL magic with modified initialization
%load_ext sql
%config SqlMagic.style = 'PRETTY'  # Fix: Change DEFAULT to PRETTY
%sql sqlite:///layoffs.db

# Function to run SQL queries directly without SQL magic
def run_sql(query):
    """Run SQL query and return results as a DataFrame"""
    return pd.read_sql_query(query, conn)

# Step 7: Run SQL queries using the function instead of magic
print("\n## Exploratory Data Analysis with SQL")
print("We'll explore the data to identify trends, patterns, and outliers in tech industry layoffs.")

# Query 1: Basic overview
print("\n### Basic Overview")
print("First, let's look at the overall structure of the data:")
overview = run_sql("SELECT * FROM layoffs_staging2 LIMIT 10")
display(overview)

# Query 2: Total number of records
total_records_df = run_sql("SELECT COUNT(*) as total_records FROM layoffs_staging2")
total_records = total_records_df.iloc[0, 0]
print(f"Total records in dataset: {total_records}")

# Query 3: Maximum layoffs in a single event
max_layoffs_df = run_sql("SELECT MAX(total_laid_off) as max_single_layoff FROM layoffs_staging2")
max_layoffs = max_layoffs_df.iloc[0, 0]
print(f"Maximum layoffs in a single event: {max_layoffs}")

# Query 4: Percentage range analysis
print("\n### Percentage Analysis")
print("Looking at the range of percentage of employees laid off:")
percentage_range = run_sql("SELECT MAX(percentage_laid_off) as max_percentage, MIN(percentage_laid_off) as min_percentage FROM layoffs_staging2 WHERE percentage_laid_off IS NOT NULL")
display(percentage_range)

# Query 5: Companies with 100% layoffs
print("\n### Companies with 100% Layoffs")
print("These companies laid off 100% of their workforce, likely indicating bankruptcy or shutdown:")
full_layoffs = run_sql("SELECT * FROM layoffs_staging2 WHERE percentage_laid_off = 1 ORDER BY funds_raised DESC")
display(full_layoffs)

# Query 6: Companies with biggest single layoff events
print("\n### Biggest Single Layoff Events")
print("Companies with the largest single layoff events:")
biggest_layoffs = run_sql("SELECT company, total_laid_off, date FROM layoffs_staging2 ORDER BY total_laid_off DESC LIMIT 10")
display(biggest_layoffs)

# Query 7: Companies with most total layoffs across all events
print("\n### Companies with Most Total Layoffs")
print("Companies with the highest total number of employees laid off across all events:")
total_by_company = run_sql("SELECT company, SUM(total_laid_off) as total_layoffs FROM layoffs_staging2 GROUP BY company ORDER BY total_layoffs DESC LIMIT 10")
display(total_by_company)

# Query 8: Layoffs by location
print("\n### Layoffs by Location")
print("Cities with the highest number of laid off employees:")
layoffs_by_location = run_sql("SELECT location, SUM(total_laid_off) as total_layoffs FROM layoffs_staging2 GROUP BY location ORDER BY total_layoffs DESC LIMIT 10")
display(layoffs_by_location)

# Query 9: Layoffs by country
print("\n### Layoffs by Country")
print("Countries with the highest number of laid off employees:")
layoffs_by_country = run_sql("SELECT country, SUM(total_laid_off) as total_layoffs FROM layoffs_staging2 WHERE country IS NOT NULL GROUP BY country ORDER BY total_layoffs DESC")
display(layoffs_by_country)

# Query 10: Layoffs by year
print("\n### Layoffs by Year")
print("Total layoffs per year:")
layoffs_by_year = run_sql("SELECT strftime('%Y', date) as year, SUM(total_laid_off) as total_layoffs FROM layoffs_staging2 GROUP BY year ORDER BY year ASC")
display(layoffs_by_year)

# Query 11: Layoffs by industry
print("\n### Layoffs by Industry")
print("Industries with the highest number of layoffs:")
layoffs_by_industry = run_sql("SELECT industry, SUM(total_laid_off) as total_layoffs FROM layoffs_staging2 GROUP BY industry ORDER BY total_layoffs DESC")
display(layoffs_by_industry)

# Query 12: Layoffs by company stage
print("\n### Layoffs by Company Stage")
print("Company stages with the highest number of layoffs:")
layoffs_by_stage = run_sql("SELECT stage, SUM(total_laid_off) as total_layoffs FROM layoffs_staging2 GROUP BY stage ORDER BY total_layoffs DESC")
display(layoffs_by_stage)

# Query 13: Top 3 companies with most layoffs per year
print("\n### Top 3 Companies with Most Layoffs Per Year")
top_companies_by_year = run_sql("""
WITH Company_Year AS (
    SELECT company, strftime('%Y', date) AS years, SUM(total_laid_off) AS total_laid_off
    FROM layoffs_staging2
    GROUP BY company, strftime('%Y', date)
),
Company_Year_Rank AS (
    SELECT company, years, total_laid_off,
    DENSE_RANK() OVER (PARTITION BY years ORDER BY total_laid_off DESC) AS ranking
    FROM Company_Year
)
SELECT company, years, total_laid_off, ranking
FROM Company_Year_Rank
WHERE ranking <= 3 AND years IS NOT NULL
ORDER BY years ASC, total_laid_off DESC
""")
display(top_companies_by_year)

# Query 14: Monthly layoffs
print("\n### Monthly Layoffs")
print("Layoffs per month:")
monthly_layoffs = run_sql("SELECT strftime('%Y-%m', date) as month, SUM(total_laid_off) as total_layoffs FROM layoffs_staging2 GROUP BY month ORDER BY month ASC")
display(monthly_layoffs)

# Query 15: Rolling total of layoffs per month
print("\n### Rolling Total of Layoffs")
print("Cumulative layoffs over time:")
rolling_layoffs = run_sql("""
WITH DATE_CTE AS (
    SELECT strftime('%Y-%m', date) as dates, SUM(total_laid_off) AS total_laid_off
    FROM layoffs_staging2
    GROUP BY dates
    ORDER BY dates ASC
)
SELECT dates, total_laid_off, SUM(total_laid_off) OVER (ORDER BY dates ASC) as rolling_total_layoffs
FROM DATE_CTE
ORDER BY dates ASC
""")
display(rolling_layoffs)

# Query 16: Average percentage laid off by industry
print("\n### Average Percentage Laid Off by Industry")
avg_percentage_by_industry = run_sql("SELECT industry, AVG(percentage_laid_off) as avg_percentage FROM layoffs_staging2 WHERE percentage_laid_off IS NOT NULL GROUP BY industry ORDER BY avg_percentage DESC")
display(avg_percentage_by_industry)

# Query 17: Relationship between funding and layoffs
print("\n### Relationship Between Funding and Layoffs")
funding_vs_layoffs = run_sql("SELECT company, funds_raised, total_laid_off, percentage_laid_off FROM layoffs_staging2 WHERE funds_raised IS NOT NULL AND percentage_laid_off IS NOT NULL ORDER BY funds_raised DESC LIMIT 20")
display(funding_vs_layoffs)

# Query 18: Quarterly analysis of layoffs
print("\n### Quarterly Analysis of Layoffs")
quarterly_layoffs = run_sql("SELECT strftime('%Y', date) || '-Q' || ((strftime('%m', date) + 2) / 3) as quarter, SUM(total_laid_off) as total_layoffs FROM layoffs_staging2 GROUP BY quarter ORDER BY quarter ASC")
display(quarterly_layoffs)

# Query 19: Companies with multiple layoff events
print("\n### Companies with Multiple Layoff Events")
multiple_layoffs = run_sql("""
WITH layoff_counts AS (
    SELECT company, COUNT(*) as layoff_events, SUM(total_laid_off) as total_layoffs
    FROM layoffs_staging2
    GROUP BY company
    HAVING COUNT(*) > 1
)
SELECT company, layoff_events, total_layoffs
FROM layoff_counts
ORDER BY layoff_events DESC, total_layoffs DESC
LIMIT 15
""")
display(multiple_layoffs)

# Query 20: Layoffs by headquarters region
print("\n### Layoffs by Headquarters Region")
regional_analysis = run_sql("""
SELECT
    CASE
        WHEN country = 'United States' AND location IN ('San Francisco', 'San Francisco Bay Area', 'Palo Alto', 'Mountain View', 'Menlo Park', 'Redwood City', 'Sunnyvale', 'San Jose', 'Santa Clara') THEN 'Silicon Valley'
        WHEN country = 'United States' AND location IN ('New York', 'New York City', 'Brooklyn', 'Manhattan') THEN 'NYC Metro'
        WHEN country = 'United States' AND location IN ('Seattle', 'Bellevue', 'Redmond') THEN 'Seattle Metro'
        WHEN country = 'United States' AND location NOT IN ('San Francisco', 'San Francisco Bay Area', 'Palo Alto', 'Mountain View', 'Menlo Park', 'Redwood City', 'Sunnyvale', 'San Jose', 'Santa Clara', 'New York', 'New York City', 'Brooklyn', 'Manhattan', 'Seattle', 'Bellevue', 'Redmond') THEN 'Other US'
        WHEN country != 'United States' THEN country
        ELSE 'Unspecified'
    END as region,
    SUM(total_laid_off) as total_layoffs
FROM layoffs_staging2
GROUP BY region
ORDER BY total_layoffs DESC
""")
display(regional_analysis)

# Step 8: Data Visualization
print("\n## Data Visualizations")
print("Now that we've analyzed the data with SQL, let's create some visualizations to better understand the patterns.")

# Plot 1: Total layoffs by year
print("\n### Total Layoffs by Year")
df_year = layoffs_by_year  # No need for .DataFrame() since it's already a DataFrame
plt.figure(figsize=(10, 6))
sns.barplot(x='year', y='total_layoffs', data=df_year)
plt.title('Total Layoffs by Year', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Employees Laid Off', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Plot 2: Top 10 companies with most layoffs
print("\n### Top 10 Companies with Most Layoffs")
df_company = total_by_company  # No need for .DataFrame()
plt.figure(figsize=(12, 8))
ax = sns.barplot(x='total_layoffs', y='company', data=df_company)
plt.title('Top 10 Companies with Most Layoffs', fontsize=16)
plt.xlabel('Total Employees Laid Off', fontsize=12)
plt.ylabel('Company', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Add value labels to the bars
for i, v in enumerate(df_company['total_layoffs']):
    ax.text(v + 100, i, f'{v:,}', va='center')
plt.show()

# Plot 3: Layoffs by industry
print("\n### Layoffs by Industry")
df_industry = layoffs_by_industry  # No need for .DataFrame()
plt.figure(figsize=(12, 10))
ax = sns.barplot(x='total_layoffs', y='industry', data=df_industry)
plt.title('Total Layoffs by Industry', fontsize=16)
plt.xlabel('Number of Employees Laid Off', fontsize=12)
plt.ylabel('Industry', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Add value labels to the bars
for i, v in enumerate(df_industry['total_layoffs']):
    ax.text(v + 100, i, f'{v:,}', va='center')
plt.show()

# Plot 4: Monthly trend of layoffs
print("\n### Monthly Trend of Layoffs")
df_monthly = monthly_layoffs  # No need for .DataFrame()
df_monthly['month'] = pd.to_datetime(df_monthly['month'] + '-01')
plt.figure(figsize=(15, 8))
plt.plot(df_monthly['month'], df_monthly['total_layoffs'], marker='o')
plt.title('Monthly Trend of Layoffs', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Employees Laid Off', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot 5: Rolling total of layoffs
print("\n### Cumulative Layoffs Over Time")
df_rolling = rolling_layoffs  # No need for .DataFrame()
df_rolling['dates'] = pd.to_datetime(df_rolling['dates'] + '-01')
plt.figure(figsize=(15, 8))
plt.plot(df_rolling['dates'], df_rolling['rolling_total_layoffs'], marker='o')
plt.title('Cumulative Layoffs Over Time', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Cumulative Number of Employees Laid Off', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot 6: Top locations for layoffs
print("\n### Top Locations for Layoffs")
df_location = layoffs_by_location  # No need for .DataFrame()
plt.figure(figsize=(12, 8))
ax = sns.barplot(x='total_layoffs', y='location', data=df_location)
plt.title('Top 10 Locations with Most Layoffs', fontsize=16)
plt.xlabel('Number of Employees Laid Off', fontsize=12)
plt.ylabel('Location', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Add value labels to the bars
for i, v in enumerate(df_location['total_layoffs']):
    ax.text(v + 100, i, f'{v:,}', va='center')
plt.show()

# Plot 7: Layoffs by company stage
print("\n### Layoffs by Company Stage")
df_stage = layoffs_by_stage  # No need for .DataFrame()
plt.figure(figsize=(12, 8))
plt.pie(df_stage['total_layoffs'], labels=df_stage['stage'], autopct='%1.1f%%',
        shadow=True, startangle=140, explode=[0.05]*len(df_stage))
plt.title('Distribution of Layoffs by Company Stage', fontsize=16)
plt.axis('equal')
plt.show()

# Plot 8: Quarterly trend of layoffs
print("\n### Quarterly Trend of Layoffs")
df_quarterly = quarterly_layoffs  # No need for .DataFrame()
plt.figure(figsize=(14, 8))
sns.barplot(x='quarter', y='total_layoffs', data=df_quarterly)
plt.title('Quarterly Trend of Layoffs', fontsize=16)
plt.xlabel('Quarter', fontsize=12)
plt.ylabel('Number of Employees Laid Off', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Plot 9: Relationship between funding and layoffs
print("\n### Relationship Between Funding and Layoffs")
# Extract non-null values
funding_data = df[['funds_raised', 'total_laid_off']].dropna()
plt.figure(figsize=(12, 8))
plt.scatter(funding_data['funds_raised'], funding_data['total_laid_off'], alpha=0.6)
plt.title('Relationship Between Funding and Layoffs', fontsize=16)
plt.xlabel('Funds Raised (Millions $)', fontsize=12)
plt.ylabel('Employees Laid Off', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xscale('log')  # Use log scale for better visualization
plt.show()

# Plot 10: Layoffs by regions (using Plotly for interactive visualization)
print("\n### Layoffs by Region (Interactive)")
df_region = regional_analysis  # No need for .DataFrame()
fig = px.pie(df_region, values='total_layoffs', names='region',
             title='Distribution of Layoffs by Region',
             hover_data=['total_layoffs'])
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()

# Plot 11: Companies with multiple layoff events
print("\n### Companies with Multiple Layoff Events")
df_multiple = multiple_layoffs  # No need for .DataFrame()
fig = px.scatter(df_multiple, x='layoff_events', y='total_layoffs',
                 text='company', size='total_layoffs',
                 title='Companies with Multiple Layoff Events',
                 labels={'layoff_events': 'Number of Layoff Events',
                         'total_layoffs': 'Total Employees Laid Off'})
fig.update_traces(textposition='top center')
fig.show()

# Step 9: Advanced Analysis
print("\n## Advanced Analysis")

# Time series decomposition
print("\n### Time Series Decomposition")
print("Analyzing the trend, seasonal, and residual components of the layoff data:")

# Prepare time series data
try:
    from statsmodels.tsa.seasonal import seasonal_decompose

    # Resample to monthly frequency and fill missing values
    df_ts = df.copy()
    df_ts['month'] = df_ts['date'].dt.to_period('M').dt.to_timestamp()
    monthly_ts = df_ts.groupby('month')['total_laid_off'].sum().reset_index().set_index('month')
    monthly_ts = monthly_ts.asfreq('MS')
    monthly_ts = monthly_ts.fillna(0)

    # Perform decomposition
    decomposition = seasonal_decompose(monthly_ts, model='additive', period=12)

    # Plot decomposition
    plt.figure(figsize=(14, 12))
    plt.subplot(411)
    plt.plot(monthly_ts, label='Observed')
    plt.legend(loc='upper left')
    plt.title('Time Series Decomposition of Monthly Layoffs')

    plt.subplot(412)
    plt.plot(decomposition.trend, label='Trend')
    plt.legend(loc='upper left')

    plt.subplot(413)
    plt.plot(decomposition.seasonal, label='Seasonal')
    plt.legend(loc='upper left')

    plt.subplot(414)
    plt.plot(decomposition.resid, label='Residual')
    plt.legend(loc='upper left')

    plt.tight_layout()
    plt.show()

    print("✅ Time series decomposition completed successfully")
except Exception as e:
    print(f"❌ Could not perform time series decomposition: {str(e)}")
    print("You may need to install statsmodels: !pip install statsmodels")

# Correlation analysis
print("\n### Correlation Analysis")
print("Analyzing correlations between numerical variables:")

# Prepare data for correlation analysis
num_cols = ['total_laid_off', 'percentage_laid_off', 'funds_raised']
corr_data = df[num_cols].dropna()

if len(corr_data) > 0:
    # Calculate correlation matrix
    corr_matrix = corr_data.corr()

    # Plot correlation heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
    plt.title('Correlation Matrix of Numerical Variables', fontsize=16)
    plt.show()

    print("Interpretation:")
    for i in range(len(num_cols)):
        for j in range(i+1, len(num_cols)):
            corr_val = corr_matrix.iloc[i, j]
            print(f"- Correlation between {num_cols[i]} and {num_cols[j]}: {corr_val:.2f}")
            if abs(corr_val) < 0.3:
                print(f"  Weak correlation")
            elif abs(corr_val) < 0.7:
                print(f"  Moderate correlation")
            else:
                print(f"  Strong correlation")
else:
    print("Not enough data with all numerical variables for correlation analysis")

# Step 10: Summary and Insights
print("\n## Summary and Insights")
print("Based on our analysis of the layoffs dataset, here are the key findings:")

# Summary calculations
total_laid_off = df['total_laid_off'].sum()
total_companies = df['company'].nunique()
avg_percentage = df['percentage_laid_off'].mean() * 100
max_single = df['total_laid_off'].max()
top_company = total_by_company['company'].iloc[0]
top_industry = layoffs_by_industry['industry'].iloc[0]

print(f"1. Total employees laid off: {total_laid_off:,}")
print(f"2. Total unique companies with layoffs: {total_companies}")
print(f"3. Average percentage of workforce laid off: {avg_percentage:.1f}%")
print(f"4. Maximum employees laid off in a single event: {max_single:,}")
print(f"5. Company with most layoffs: {top_company}")
print(f"6. Industry with most layoffs: {top_industry}")

# Year-over-year analysis
if len(df_year) >= 2:
    yoy_change = ((df_year['total_layoffs'].iloc[-1] / df_year['total_layoffs'].iloc[-2]) - 1) * 100
    print(f"7. Year-over-year change in layoffs (last two years): {yoy_change:.1f}%")

print("\n### Key Insights:")
print("- Silicon Valley and major tech hubs show the highest concentration of layoffs")
print("- Companies with significant funding aren't immune to layoffs")
print("- Several companies had multiple rounds of layoffs, indicating ongoing restructuring")
print("- There are clear seasonal patterns in layoff announcements")
print("- Economic factors such as interest rates and market conditions appear to influence layoff trends")

print("\n### Recommendations for Further Analysis:")
print("1. Incorporate macroeconomic indicators to better understand external influences")
print("2. Analyze stock performance before and after layoff announcements")
print("3. Investigate the relationship between company age and layoff patterns")
print("4. Compare layoff patterns with hiring trends in the same period")
print("5. Conduct sentiment analysis on news and social media related to major layoff events")

# Step 11: Save results to Google Drive
print("\n## Saving Results")
try:
    # Save the processed dataframe
    df.to_csv('/content/drive/My Drive/Data_Analyst/Portfolio_Projects/Projects/SQL/layoffs_processed.csv', index=False)

    # Save a summary report
    summary_df = pd.DataFrame({
        'Metric': ['Total Employees Laid Off', 'Unique Companies', 'Average Layoff Percentage',
                  'Max Single Layoff', 'Top Company', 'Top Industry'],
        'Value': [f"{total_laid_off:,}", f"{total_companies}", f"{avg_percentage:.1f}%",
                 f"{max_single:,}", f"{top_company}", f"{top_industry}"]
    })

    summary_df.to_csv('/content/drive/My Drive/Data_Analyst/Portfolio_Projects/Projects/SQL/layoffs_summary.csv', index=False)
    print("✅ Results saved to Google Drive")
except Exception as e:
    print(f"❌ Could not save results to Google Drive: {str(e)}")
    print("Results saved locally instead")
    # Save locally as fallback
    df.to_csv('layoffs_processed.csv', index=False)

print("\n## Analysis Complete")
print("This comprehensive analysis has explored the layoffs dataset from multiple angles.")
```
