```python
# ============================================================================
# COMPREHENSIVE DATA ANALYSIS SCRIPT FOR GOOGLE COLAB
# ============================================================================
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import os
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')
from google.colab import drive
drive.mount('/content/drive')

# Set display and visualization options
pd.set_option('display.float_format', lambda x: '%.2f' % x)
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")
plt.style.use('seaborn-v0_8-whitegrid')

print("=" * 80)
print("DATA ANALYSIS SCRIPT EXECUTION STARTED")
print("=" * 80)
print(f"Execution started at: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print("-" * 80)

# Define input and output file paths
input_file = '/content/drive/My Drive/Data_Analyst/Portfolio_Projects/Projects/Python/Calls_Dataset_After_Cleaning.xlsx'
output_folder = '/content/drive/My Drive/Data_Analyst/Portfolio_Projects/Projects/Python/'
output_file = os.path.join(output_folder, 'Comprehensive_Analysis_Results.xlsx')
plots_folder = os.path.join(output_folder, 'Analysis_Plots')

# Create plots folder if it doesn't exist
if not os.path.exists(plots_folder):
    os.makedirs(plots_folder)
    print(f"Created plots folder at: {plots_folder}")

# ============================================================================
# STEP 1: DATA LOADING AND INITIAL EXPLORATION
# ============================================================================
print("\n1. LOADING DATA AND INITIAL EXPLORATION")
print("-" * 80)
print(f"Loading data from: {input_file}")

# Load the data
try:
    df = pd.read_excel(input_file)
    print(f"âœ“ Successfully loaded data with {df.shape[0]} rows and {df.shape[1]} columns")
except Exception as e:
    print(f"âŒ Error loading data: {str(e)}")
    raise

# Display the first few rows of the dataset
print("\n1.1. Preview of the dataset (first 5 rows):")
print(df.head())

# Print column names to identify the actual columns in the dataset
print("\n1.2. Available columns in the dataset:")
print(df.columns.tolist())

# Get data types of each column
print("\n1.3. Data types of each column:")
print(df.dtypes)

# ============================================================================
# STEP 2: DATA UNDERSTANDING AND SUMMARY STATISTICS
# ============================================================================
print("\n2. DATA UNDERSTANDING AND SUMMARY STATISTICS")
print("-" * 80)
print("This step provides insights into the structure, distribution, and overall characteristics of the data.")

# Dataset information
print("\n2.1. DataFrame Information (structure, data types, non-null counts):")
df.info()

# Summary statistics for numerical columns
print("\n2.2. Summary Statistics for numerical columns:")
print("This provides count, mean, std, min, 25%, 50%, 75%, max values.")
print(df.describe())

# Summary statistics for categorical columns (if any)
categorical_cols = df.select_dtypes(include=['object']).columns
if len(categorical_cols) > 0:
    print("\n2.3. Summary for categorical columns:")
    for col in categorical_cols:
        print(f"\nValue counts for {col}:")
        print(df[col].value_counts().head(10))
        print(f"Number of unique values: {df[col].nunique()}")

# Check for missing values
print("\n2.4. Missing Values Analysis:")
missing_values = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100
missing_df = pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percent})
print(missing_df[missing_df['Missing Values'] > 0])

if missing_df['Missing Values'].sum() == 0:
    print("âœ“ No missing values found in the dataset")

# Count unique values in each column
print("\n2.5. Count of unique values in each column:")
print(df.nunique())

# ============================================================================
# STEP 3: IDENTIFY NUMERIC AND CATEGORICAL COLUMNS FOR ANALYSIS
# ============================================================================
print("\n3. IDENTIFYING COLUMN TYPES FOR ANALYSIS")
print("-" * 80)
print("Separating numeric and categorical columns for appropriate analysis methods.")

# Identify numeric columns (exclude datetime types)
numeric_df = df.select_dtypes(include=['float64', 'int64'])
numeric_cols = numeric_df.columns.tolist()

# Identify categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

# Identify datetime columns (if any)
datetime_cols = df.select_dtypes(include=['datetime64']).columns.tolist()

print(f"\n3.1. Identified {len(numeric_cols)} numeric columns:")
print(numeric_cols)

print(f"\n3.2. Identified {len(categorical_cols)} categorical columns:")
print(categorical_cols)

if datetime_cols:
    print(f"\n3.3. Identified {len(datetime_cols)} datetime columns:")
    print(datetime_cols)

# ============================================================================
# STEP 4: UNIVARIATE ANALYSIS - NUMERIC COLUMNS
# ============================================================================
print("\n4. UNIVARIATE ANALYSIS - NUMERIC COLUMNS")
print("-" * 80)
print("Analyzing the distribution and characteristics of individual numeric variables.")

if len(numeric_cols) > 0:
    # Basic statistics for numeric columns
    print("\n4.1. Detailed statistics for numeric columns:")
    numeric_stats = numeric_df.describe(percentiles=[.05, .25, .5, .75, .95]).T
    numeric_stats['IQR'] = numeric_stats['75%'] - numeric_stats['25%']
    numeric_stats['skew'] = numeric_df.skew()
    numeric_stats['kurtosis'] = numeric_df.kurtosis()
    print(numeric_stats)

    # Distribution plots for numeric columns
    print("\n4.2. Creating distribution plots for numeric columns...")

    # Histograms
    plt.figure(figsize=(15, len(numeric_cols) * 3))
    for i, col in enumerate(numeric_cols, 1):
        plt.subplot(len(numeric_cols), 1, i)
        sns.histplot(df[col].dropna(), kde=True)
        plt.title(f'Distribution of {col}')
        plt.tight_layout()
    plt.savefig(os.path.join(plots_folder, 'histograms.png'))
    plt.show()
    print("âœ“ Created histogram distributions for all numeric columns")

    # Box plots
    plt.figure(figsize=(15, 10))
    plt.title('Boxplots of All Numeric Variables')
    numeric_df.boxplot(vert=False, figsize=(15, 10))
    plt.tight_layout()
    plt.savefig(os.path.join(plots_folder, 'boxplots.png'))
    plt.show()
    print("âœ“ Created boxplots for all numeric columns")

    # Check for outliers using IQR method
    print("\n4.3. Outlier Detection using IQR method:")
    for col in numeric_cols:
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - (1.5 * iqr)
        upper_bound = q3 + (1.5 * iqr)
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
        if len(outliers) > 0:
            print(f"Column '{col}' has {len(outliers)} outliers ({(len(outliers)/len(df))*100:.2f}% of data)")
            print(f"  - Outlier range: < {lower_bound:.2f} or > {upper_bound:.2f}")
else:
    print("No numeric columns found for univariate analysis")

# ============================================================================
# STEP 5: UNIVARIATE ANALYSIS - CATEGORICAL COLUMNS
# ============================================================================
print("\n5. UNIVARIATE ANALYSIS - CATEGORICAL COLUMNS")
print("-" * 80)
print("Analyzing the distribution and frequencies of categorical variables.")

if len(categorical_cols) > 0:
    # Frequency distribution for top categorical values
    print("\n5.1. Frequency distribution for categorical columns:")
    for col in categorical_cols:
        value_counts = df[col].value_counts().head(10)
        print(f"\nTop 10 values for '{col}':")
        print(value_counts)

        # Bar plots for categorical columns
        plt.figure(figsize=(12, 6))
        sns.countplot(y=col, data=df, order=df[col].value_counts().iloc[:10].index)
        plt.title(f'Top 10 values in {col}')
        plt.tight_layout()
        plt.savefig(os.path.join(plots_folder, f'barplot_{col}.png'))
        plt.show()
        print(f"âœ“ Created bar plot for '{col}'")

    # Identify the main categorical column for grouping
    if 'Continent' in categorical_cols:
        main_cat_col = 'Continent'
    else:
        # Use the categorical column with the fewest unique values for more meaningful analysis
        unique_counts = {col: df[col].nunique() for col in categorical_cols}
        main_cat_col = min(unique_counts, key=unique_counts.get)

    print(f"\n5.2. Selected '{main_cat_col}' as the main categorical column for grouping analysis")
    print(f"It has {df[main_cat_col].nunique()} unique values")
else:
    main_cat_col = None
    print("No categorical columns found for univariate analysis")

# ============================================================================
# STEP 6: BIVARIATE ANALYSIS - CORRELATIONS AND RELATIONSHIPS
# ============================================================================
print("\n6. BIVARIATE ANALYSIS - CORRELATIONS AND RELATIONSHIPS")
print("-" * 80)
print("Examining relationships between pairs of variables.")

if len(numeric_cols) > 1:
    # Correlation analysis
    print("\n6.1. Correlation Matrix:")
    correlation = numeric_df.corr()
    print(correlation)

    # Visualize correlation matrix
    plt.figure(figsize=(max(12, len(numeric_cols)), max(10, len(numeric_cols))))
    sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
    plt.title('Correlation Matrix')
    plt.tight_layout()
    plt.savefig(os.path.join(plots_folder, 'correlation_heatmap.png'))
    plt.show()
    print("âœ“ Created correlation heatmap")

    # Identify strong correlations
    print("\n6.2. Strong correlations (|r| > 0.7):")
    strong_corr = (correlation.abs() > 0.7) & (correlation.abs() < 1.0)
    if strong_corr.any().any():
        pairs = []
        for col in correlation.columns:
            for idx in correlation.index:
                if strong_corr.loc[idx, col] and (col, idx) not in pairs and (idx, col) not in pairs and idx != col:
                    pairs.append((idx, col))
                    print(f"Strong correlation between '{idx}' and '{col}': {correlation.loc[idx, col]:.2f}")
    else:
        print("No strong correlations found")

    # Scatter plots for correlated variables
    if len(pairs) > 0:
        print("\n6.3. Creating scatter plots for correlated variables...")
        for pair in pairs[:5]:  # Limit to 5 pairs to avoid too many plots
            plt.figure(figsize=(10, 6))
            sns.scatterplot(x=pair[0], y=pair[1], data=df)
            plt.title(f'Scatter plot: {pair[0]} vs {pair[1]}')
            plt.tight_layout()
            plt.savefig(os.path.join(plots_folder, f'scatter_{pair[0]}_{pair[1]}.png'))
            plt.show()
            print(f"âœ“ Created scatter plot for {pair[0]} vs {pair[1]}")
else:
    print("Not enough numeric columns for correlation analysis")

# Analyze categorical vs numeric relationships
if main_cat_col and len(numeric_cols) > 0:
    print("\n6.4. Analyzing the relationship between categorical and numeric variables:")
    # Find the top 5 categorical values by frequency
    top_cats = df[main_cat_col].value_counts().head(5).index

    # For each numeric column, compare distributions across categories
    for col in numeric_cols[:3]:  # Limit to first 3 numeric columns
        # Box plots for numeric variable by category
        plt.figure(figsize=(12, 6))
        sns.boxplot(x=main_cat_col, y=col, data=df[df[main_cat_col].isin(top_cats)])
        plt.title(f'{col} by {main_cat_col}')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.savefig(os.path.join(plots_folder, f'boxplot_{col}_by_{main_cat_col}.png'))
        plt.show()
        print(f"âœ“ Created boxplot for {col} by {main_cat_col}")

        # Bar plot for mean of numeric variable by category
        plt.figure(figsize=(12, 6))
        # Only use the numeric column with the categorical column to avoid errors
        mean_by_cat = df.groupby(main_cat_col)[col].mean().sort_values(ascending=False).head(10)
        mean_by_cat.plot(kind='bar')
        plt.title(f'Mean {col} by {main_cat_col}')
        plt.ylabel(f'Mean {col}')
        plt.tight_layout()
        plt.savefig(os.path.join(plots_folder, f'barplot_mean_{col}_by_{main_cat_col}.png'))
        plt.show()
        print(f"âœ“ Created bar plot for mean {col} by {main_cat_col}")

# ============================================================================
# STEP 7: TIME SERIES ANALYSIS (if applicable)
# ============================================================================
print("\n7. TIME SERIES ANALYSIS (if applicable)")
print("-" * 80)
print("Analyzing patterns over time if time-related data is available.")

# Check for date columns or columns with year patterns
date_cols = datetime_cols.copy() if datetime_cols else []
year_pattern_cols = [col for col in numeric_cols if any(str(year) in str(col) for year in range(1900, 2023))]

if date_cols or year_pattern_cols:
    print("\n7.1. Identified potential time-related columns:")
    if date_cols:
        print(f"Date columns: {date_cols}")
    if year_pattern_cols:
        print(f"Year pattern columns: {year_pattern_cols}")

    # Time series analysis with year pattern columns
    if year_pattern_cols:
        print("\n7.2. Time Series Analysis with year pattern columns:")

        # Prepare data for time series plot
        if main_cat_col:
            # Group by categorical column and calculate mean for each year
            ts_data = df.groupby(main_cat_col)[year_pattern_cols].mean()

            # Sort categories by the most recent year's values
            if len(year_pattern_cols) > 0:
                most_recent_year = max(year_pattern_cols)
                ts_data_sorted = ts_data.sort_values(by=most_recent_year, ascending=False)

                # Transpose for time series plotting
                ts_plot_data = ts_data_sorted.transpose()

                # Plot time series
                plt.figure(figsize=(14, 8))
                for category in ts_plot_data.columns[:5]:  # Limit to top 5 categories
                    plt.plot(ts_plot_data.index, ts_plot_data[category], marker='o', label=category)

                plt.title(f'Time Series by {main_cat_col}')
                plt.xlabel('Year')
                plt.ylabel('Value')
                plt.legend()
                plt.grid(True)
                plt.tight_layout()
                plt.savefig(os.path.join(plots_folder, 'time_series_plot.png'))
                plt.show()
                print(f"âœ“ Created time series plot by {main_cat_col}")

                # Growth rate analysis
                if len(year_pattern_cols) > 1:
                    print("\n7.3. Growth Rate Analysis:")
                    # Calculate year-over-year growth rates
                    growth_rates = pd.DataFrame()

                    for i in range(1, len(year_pattern_cols)):
                        prev_year = year_pattern_cols[i-1]
                        curr_year = year_pattern_cols[i]
                        col_name = f'Growth_{prev_year}_to_{curr_year}'
                        growth_rates[col_name] = (ts_data[curr_year] - ts_data[prev_year]) / ts_data[prev_year] * 100

                    print(growth_rates.head())

                    # Plot average growth rate by category
                    avg_growth = growth_rates.mean(axis=1).sort_values(ascending=False)
                    plt.figure(figsize=(12, 6))
                    avg_growth.head(10).plot(kind='bar')
                    plt.title(f'Average Growth Rate by {main_cat_col}')
                    plt.ylabel('Average Growth Rate (%)')
                    plt.tight_layout()
                    plt.savefig(os.path.join(plots_folder, 'average_growth_rate.png'))
                    plt.show()
                    print(f"âœ“ Created average growth rate plot by {main_cat_col}")
else:
    print("No time-related columns identified for time series analysis")

# ============================================================================
# STEP 8: ADVANCED STATISTICAL ANALYSIS
# ============================================================================
print("\n8. ADVANCED STATISTICAL ANALYSIS")
print("-" * 80)
print("Performing more detailed statistical tests and analysis.")

if len(numeric_cols) > 0:
    # Distribution normality test
    print("\n8.1. Normality Test for Numeric Variables:")
    for col in numeric_cols:
        if df[col].count() > 5:  # Need at least 8 observations for Shapiro-Wilk test
            stat, p = stats.shapiro(df[col].dropna())
            print(f"Shapiro-Wilk Test for '{col}': stat={stat:.4f}, p-value={p:.4f}")
            if p > 0.05:
                print(f"  âœ“ '{col}' appears to be normally distributed (p > 0.05)")
            else:
                print(f"  âœ— '{col}' does not appear to be normally distributed (p <= 0.05)")

    # Z-score analysis for outliers
    print("\n8.2. Z-score Analysis for Outliers:")
    for col in numeric_cols:
        z_scores = stats.zscore(df[col].dropna())
        outliers = len([x for x in z_scores if abs(x) > 3])
        if outliers > 0:
            print(f"Column '{col}' has {outliers} outliers based on z-score > 3 ({(outliers/len(df))*100:.2f}% of data)")

# Group comparison if categorical and numeric columns exist
if main_cat_col and len(numeric_cols) > 0:
    print("\n8.3. Group Comparison Analysis:")
    # Get top categories by frequency
    top_cats = df[main_cat_col].value_counts().head(3).index.tolist()

    # Filter data for those categories
    filtered_df = df[df[main_cat_col].isin(top_cats)]

    # For each numeric column, compare means across groups
    for col in numeric_cols[:3]:  # Limit to first 3 numeric columns
        print(f"\nAnalysis of '{col}' across '{main_cat_col}' groups:")

        # Group statistics
        group_stats = filtered_df.groupby(main_cat_col)[col].agg(['count', 'mean', 'std', 'min', 'max'])
        print(group_stats)

        # If we have at least 2 groups, perform ANOVA
        if len(top_cats) >= 2 and all(filtered_df.groupby(main_cat_col)[col].count() > 5):
            try:
                groups = [filtered_df[filtered_df[main_cat_col] == cat][col].dropna() for cat in top_cats]
                f_stat, p_val = stats.f_oneway(*groups)
                print(f"One-way ANOVA: F={f_stat:.4f}, p-value={p_val:.4f}")

                if p_val < 0.05:
                    print(f"  âœ“ There is a statistically significant difference in '{col}' among '{main_cat_col}' groups (p < 0.05)")
                else:
                    print(f"  âœ— No statistically significant difference in '{col}' among '{main_cat_col}' groups (p >= 0.05)")
            except:
                print(f"  Could not perform ANOVA test for '{col}'")

# ============================================================================
# STEP 9: SUMMARY AND INSIGHTS
# ============================================================================
print("\n9. SUMMARY AND INSIGHTS")
print("-" * 80)
print("Summarizing key findings from the analysis.")

# Dataset summary
print(f"\n9.1. Dataset Overview:")
print(f"  â€¢ Number of records: {df.shape[0]}")
print(f"  â€¢ Number of columns: {df.shape[1]}")
print(f"  â€¢ Number of numeric columns: {len(numeric_cols)}")
print(f"  â€¢ Number of categorical columns: {len(categorical_cols)}")
if 'missing_df' in locals():
    missing_count = missing_df[missing_df['Missing Values'] > 0].shape[0]
    print(f"  â€¢ Columns with missing values: {missing_count}")

# Top insights based on analysis
print("\n9.2. Key Insights:")
if len(numeric_cols) > 0:
    # Identify column with highest variability (coefficient of variation)
    cv = numeric_df.std() / numeric_df.mean()
    highest_var_col = cv.dropna().idxmax()
    print(f"  â€¢ Highest variability: '{highest_var_col}' has the highest relative variability (CV = {cv[highest_var_col]:.2f})")

    # Identify columns with outliers
    print("  â€¢ Outliers detected in the following columns:")
    outlier_cols = []
    for col in numeric_cols:
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - (1.5 * iqr)
        upper_bound = q3 + (1.5 * iqr)
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
        if len(outliers) > 0:
            outlier_cols.append(col)
            print(f"    - {col}: {len(outliers)} outliers ({(len(outliers)/len(df))*100:.2f}% of data)")

    if not outlier_cols:
        print("    - No significant outliers detected")

if 'correlation' in locals():
    # Strongest correlations
    strong_positive = correlation.unstack().sort_values(ascending=False)
    strong_positive = strong_positive[(strong_positive < 1.0) & (strong_positive > 0.7)]

    strong_negative = correlation.unstack().sort_values()
    strong_negative = strong_negative[(strong_negative > -1.0) & (strong_negative < -0.7)]

    if len(strong_positive) > 0:
        print("  â€¢ Strongest positive correlations:")
        for idx, val in strong_positive[:3].items():
            if idx[0] != idx[1]:  # Skip self-correlations
                print(f"    - {idx[0]} and {idx[1]}: {val:.2f}")

    if len(strong_negative) > 0:
        print("  â€¢ Strongest negative correlations:")
        for idx, val in strong_negative[:3].items():
            if idx[0] != idx[1]:  # Skip self-correlations
                print(f"    - {idx[0]} and {idx[1]}: {val:.2f}")

if main_cat_col:
    print(f"  â€¢ Most frequent {main_cat_col}: {df[main_cat_col].value_counts().idxmax()} ({df[main_cat_col].value_counts().max()} occurrences)")

    # Fixed section for group statistics
    if 'group_stats' in locals():
        print(f"  â€¢ Group with highest mean value:")
        for col in numeric_cols[:2]:
            try:
                # Safely get the group with the maximum mean value
                max_group = group_stats['mean'].idxmax()
                max_value = group_stats.loc[max_group, 'mean']
                print(f"    - For {col}: {max_group} (Mean: {max_value:.2f})")
            except:
                print(f"    - Cannot determine group with highest mean for {col}")

if 'ts_data_sorted' in locals():
    print("  â€¢ Time series insights:")
    print(f"    - Categories with the highest values in the most recent period ({most_recent_year}):")
    for cat in ts_data_sorted.index[:3]:
        print(f"      * {cat}: {ts_data_sorted.loc[cat, most_recent_year]:.2f}")

    if 'avg_growth' in locals():
        print(f"    - Categories with the highest average growth rates:")
        for cat in avg_growth.head(3).index:
            print(f"      * {cat}: {avg_growth[cat]:.2f}%")

# ============================================================================
# STEP 10: SAVE RESULTS TO EXCEL
# ============================================================================
print("\n10. SAVING RESULTS")
print("-" * 80)
print(f"Saving analysis results to: {output_file}")

# Create an Excel writer
with pd.ExcelWriter(output_file, engine='openpyxl') as writer:
    # Basic dataframe
    df.head(100).to_excel(writer, sheet_name='Data_Sample', index=False)

    # Basic statistics - only for numeric columns to avoid errors
    if not numeric_df.empty:
        numeric_df.describe().to_excel(writer, sheet_name='Summary_Stats')

    # Missing values
    missing_df.to_excel(writer, sheet_name='Missing_Values')

    # Correlation matrix
    if 'correlation' in locals() and not correlation.empty:
        correlation.to_excel(writer, sheet_name='Correlation')

    # Group statistics
    if main_cat_col and not numeric_df.empty:
        group_means = df.groupby(main_cat_col)[numeric_cols].mean()
        group_means.to_excel(writer, sheet_name='Group_Means')

    # Time series data
    if 'ts_data_sorted' in locals() and not ts_data_sorted.empty:
        ts_data_sorted.to_excel(writer, sheet_name='Time_Series_Data')
        ts_plot_data.to_excel(writer, sheet_name='Time_Series_Plot_Data')

    # Growth rates
    if 'growth_rates' in locals() and not growth_rates.empty:
        growth_rates.to_excel(writer, sheet_name='Growth_Rates')

    # Detailed numeric statistics
    if 'numeric_stats' in locals() and not numeric_stats.empty:
        numeric_stats.to_excel(writer, sheet_name='Detailed_Stats')

print(f"âœ“ Results saved to: {output_file}")
print(f"âœ“ Plots saved to: {plots_folder}")

print("=" * 80)
print("DATA ANALYSIS SCRIPT EXECUTION COMPLETED")
print(f"Execution finished at: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print("=" * 80)
```
